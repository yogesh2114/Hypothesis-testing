```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
lab=pd.read_csv('LabTAT.csv')
lab
```

```python
lab_pd=pd.DataFrame(lab)
lab_pd
```

```python
lab_pd.rename(columns={'Laboratory 1':'lab1','Laboratory 2':'lab2','Laboratory 3':'lab3','Laboratory 4':'lab4'},inplace=True)
```

```python
#  here we have 4 sample means like Laboratory 1,2,3,4 and 
#the decision is based on a single variable which is TAT (Turn Around Time) 
 #so we have applied the Anova oneway test because we have dealed with more than
#two sample means and single variable
#Anova testalso called as F - statistic
# now we have to define whats is the null and alternative hypothesis
 null hypothesis : all four laboratory reports are same there is no differeence
 alternative hypothesis : atlesat one laboratory report is different from others   
 # if p value > alpha(0.05) then accept null hypothesis
# if p value < alpha(0.05) then accept alternative hypothesis    
```

```python
annova=stats.f_oneway(lab_1.lab1,lab_1.lab2,lab_1.lab3,lab_1.lab4)
annova
```

```python
annova_pvalues=annova[1]
annova_pvalues
```

```python
so we have p value < alpha (0.05) now we accept alternative hypothesis
```
